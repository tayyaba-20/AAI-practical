```python
class HierarchicalPlanner:
    def __init__(self):
        self.tasks = []  # List to hold tasks at different levels

    def add_task(self, task_name, subtasks=None):
        self.tasks.append({"task": task_name, "subtasks": subtasks or []})

    def execute_plan(self):
        print("Executing Hierarchical Plan:")
        for task in self.tasks:
            self.execute_task(task)

    def execute_task(self, task):
        print(f"Executing Task: {task['task']}")
        for subtask in task['subtasks']:
            print(f"  Subtask: {subtask}")

    def visualize_plan(self):
        import networkx as nx
        import matplotlib.pyplot as plt

        G = nx.DiGraph()
        for task in self.tasks:
            G.add_node(task['task'])
            for subtask in task['subtasks']:
                G.add_edge(task['task'], subtask)

        pos = nx.spring_layout(G)
        plt.figure(figsize=(8, 5))
        nx.draw(G, pos, with_labels=True, node_color='lightblue', edge_color='gray',
                node_size=2000, font_size=12)
        plt.title("Hierarchical Planning Graph")
        plt.show()

# Example Usage
planner = HierarchicalPlanner()
planner.add_task("Plan a Trip", ["Book Flight", "Book Hotel", "Pack Bags"])
planner.add_task("Pack Bags", ["Clothes", "Toiletries", "Documents"])
planner.execute_plan()
planner.visualize_plan()
```
